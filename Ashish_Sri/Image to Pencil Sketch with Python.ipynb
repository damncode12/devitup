# Import Libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# Reading Image

```python
img = cv2.imread(r"C:\Users\HP\Videos\Captures\Earth.jpg")
img
```

# Plotting Image

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.axis("off")
plt.show()
```

# Converting Image to Grayscale Image

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(7,7))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.show()
```

# Inverting Image

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(7,7))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.show()
```

# Blurring the Image

```python
img_blur=cv2.GaussianBlur(img_invert,(115,115),0)
plt.imshow(img_blur)
```

# Smoothing the Image

```python
img_smoothing = cv2.GaussianBlur(img_invert, (25, 25),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.show()
```

# Final Sketch of Image

```python
finalsketch = cv2.divide(img_gray, 250 - img_smoothing, scale=250)
plt.figure(figsize=(8,8))
plt.imshow(finalsketch,cmap="gray")
plt.axis("off")
plt.show()
```

# Comparison of Original and Pencil Sketch Image

```python
plt.figure(figsize=(20,20))
plt.subplot(1,3,1)
plt.imshow(img)
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(finalsketch,cmap="gray")
plt.axis("off")
plt.show()
```
